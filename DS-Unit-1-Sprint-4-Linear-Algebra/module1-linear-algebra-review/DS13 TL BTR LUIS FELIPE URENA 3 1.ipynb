# Part 3 - Square Matrices

## 3.1 Find $IG$ (be sure to show your work) 

\begin{align}
G= 
\begin{bmatrix}
12 & 11 \\
7 & 10 
\end{bmatrix}
\end{align}

```
#finding the identity of the array by multiplying it to the inverse

g = np.array([[12,11],
              [7,10]])
g_inv = np.linalg.inv(g)

IG = g * g_inv

IG
```

## 3.2 Find $|H|$ and then find $|J|$.

\begin{align}
H= 
\begin{bmatrix}
12 & 11 \\
7 & 10 
\end{bmatrix}
\qquad
J= 
\begin{bmatrix}
0 & 1 & 2 \\
7 & 10 & 4 \\
3 & 2 & 0
\end{bmatrix}
\end{align}


```
h = np.array([[12,11],
              [7,10]])
j = np.array([[0,1,2],
              [7,10,4],
              [3,2,0]])

h_det = np.linalg.det(h)
j_det = np.linalg.det(j)

print(h_det)
print(j_det)
```

## 3.3 Find $H^{-1}$ and then find $J^{-1}$

```
h_inv = np.linalg.inv(h)
j_inv = np.linalg.inv(j)

print(h_inv)
print(j_inv)
```
